```julia
function detrend(y,n)
    # y is the signal with drift while n is the order of polynomial fitting
    L= length(y);
    x = Vector(1:L);
    M = zeros(L,n+1);
    
    for i=1:(n+1)
        M[:,i] = x.^(i-1);
    end
    
    w =inv(M'*M)*M'*y;
    y = y-M * w;
    
    return y;
end
```

```julia
x = Vector(1:1000)
y = Vector(sin.(x.*1.1)+x.^0.4);
```

```julia
y_new = detrend(y,5);
using Plots
plot!(plot(y),y_new)
```
